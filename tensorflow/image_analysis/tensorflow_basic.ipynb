### LinearRegression : 선형 회귀

= cost가 가장 작은 weight을 구하는 것이 목적


<용어>    
Hypothesis(가설) :예측(가설)방정식 : H(x)

Cost(비용) : Loss(손실), 비용함수 

Gradient Descent Algorithm (경사하강법) : 비용함수를 미분 


```python
# 예측(가설) 방정식 : h(x) : w*x
# x : 입력값, y : 답   
# 비용(cost)함수 
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [1, 2, 4]
def cost(x,y,w):
    c = 0
    for i in range(len(x)):
        hx = w * x[i]
        loss = (hx- y[i])**2 # 비용함수 
        c += loss 
    return c/len(x)
cost(x, y, -1) # 18.6
cost(x, y, 0)  # 4.6
cost(x, y, 1)  # 0.0
cost(x, y, 2)  # 4.6
cost(x, y, 3)  # 18.6

for i in range(-30, 50):
    w = i/10
    c = cost(x,y,w)
    print(w,c)
    plt.plot(w,c,'ro')
plt.xlabel('weight')
plt.ylabel('cost')
plt.title('cost function')
plt.show()
```

```python
# Gradient Descent Algorithm(경사하강법) : 비용함수를 미분 
def gradient_descent(x,y,w) :
    grad = 0
    for i in range(len(x)):
        hx = w * x[i]
        loss = (hx - y[i])**2 # 비용함수 
        loss_grad = (hx - y[i])*x[i]   # 비용함수의 미분 
        grad += loss_grad
    return grad/len(x)

x = [1, 2, 3]
y = [1, 2, 3] 
w = 10        # 임의의 값으로 일단 시작 
old = 100
for i in range(100) : 
    c = cost(x, y, w)
    grad = gradient_descent(x, y, w)
    w = w - grad*0.1 # Learning Rate : 0.1
    print(i,c,w,grad)
    if c >= old and abs(c - old) < 1.0e-15 :
        break 
    old = c
    plt.plot((0,5), (0, 5*w))
        
print('weight=', w)

plt.plot(x, y, 'ro')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

### 미분 
미분 : 순간 변화량, 기울기 

x축으로 1만큼 움직였을 떄 y축으로 움직인 거리 

y = 3               --->      y' = 0

y = 2*x             --->      y'= 2   (일차함수의 미분)

y = x**2          --->     y'= 2*x (이차함수의 미분)

y = (x + 1)**2    --->     y'= 2*(x + 1) 

```python
# Linear_Regression.py
# Using tensorflow

import tensorflow as tf
tf.set_random_seed(777)

# X and Y data
x_train = [1,2,3]
y_train = [1,2,3]
# y_train = [4,7,10]

W = tf.Variable(tf.random_normal([1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = x_train * W + b

cost = tf.reduce_mean(tf.square(hypothesis - y_train ))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

# start learning
for step in range(2001):
    sess.run(train)
    if step % 20 == 0 :
        # 다 볼 수 없으니까 step을 20번에 1번씩 보도록 % 20
        print(step,sess.run(cost),sess.run(W),sess.run(b))

# Test(검증) 단계

# Predict(예측) 단계
Weight = sess.run(W)
bias = sess.run(b)
print('Weight:', Weight)
print('bias:', bias)

x_train = [4, 5, 6]
hypothesis2 = x_train * W + b
pred = sess.run(hypothesis2)
print('predict:', pred)
```

```python
# multi-variable linear regression 
# 입력값 : X가 두개 이상 

# 입력데이터
import tensorflow as tf
x1_data = [73.,93.,89.,96.,73.]
x2_data = [80.,88.,91.,98.,66.]
x3_data = [75.,93.,90.,100.,70.]

y_data = [152.,185.,180.,196.,142.]

x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)

# Bulid Graph 
W1 = tf.Variable(tf.random_normal([1]), name='weight1')
W2 = tf.Variable(tf.random_normal([1]), name='weight2')
W3 = tf.Variable(tf.random_normal([1]), name='weight3')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x1*W1 + x2*W2 + x3*W3 + b

# 비용함수 
cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

# 학습시작(start training)
for step in range(100001):
    cost_val, W1_val,W2_val,W3_val, b_val, _ = \
        sess.run([cost, W1, W2, W3, b, train],
                 feed_dict={ x1:x1_data,x2:x2_data,x3:x3_data, Y:y_data})
    if step % 20 == 0:
        print(step, cost_val, W1_val,W2_val,W3_val, b_val)

# 예측단계(predict) : test model
x1_data = [73.,93.,89.,96.,73.]
x2_data = [80.,88.,91.,98.,66.]
x3_data = [75.,93.,90.,100.,70.]

print(sess.run(hypothesis, feed_dict = {x1:x1_data,x2:x2_data,x3:x3_data}))
# y_data = [152.,185.,180.,196.,142.]
# [150.79488 184.97044 180.28032 197.34216 141.10304] : 20001
# [151.16731 184.78398 180.29572 196.88875 141.5592 ] : 40001
# [151.45259 184.67514 180.81311 196.03458 141.92386] : 100001
```

### Boston 시의 주택가격 예측 Linear Regression 구현

```python
# Boston 시의 주택가격을 예측하는 tensorflow 프로그램을
# 작성하세요

# 9개의 데이터를 사용하여 주택가격을 예측하는 
# multi-variable linear regression 구현

import tensorflow as tf
import numpy as np
tf.set_random_seed(777)

xy = np.loadtxt('boston_train.csv', delimiter=',',skiprows=1, dtype=np.float32)
x = xy[:, 0:-1]
y = xy[:, [-1]]
print(x.shape, y.shape)

# print(x)
# (x = (400, 9)) x (w = (9,1)) =  y = (400,1)

# train using placeholder
X = tf.placeholder(tf.float32, shape=[None, 9])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([9, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = tf.matmul(X,W) + b 

cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-6)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer()) 

# start training
for step in range(1000):
    cost_val, W_val, _ = sess.run([cost, W, train],
                                          feed_dict={X:x, Y:y})
    if step % 10 == 0:
        print(step, cost_val, W_val)
```
