```python
import json
geo_path = 'Seoul2.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
# 데이터 정리 
import numpy as np
import pandas as pd
drainage_systemt = pd.read_csv('서울시_하수도_및_부대시설_현황_통계.csv', 
                               header=1, encoding='euc-kr', thousands=',')

drainage_systemtt = drainage_systemt[['자치구', '맨홀(개소)', '빗물받이(개소)']]
drainage_systemt = drainage_systemt[1:29]
drainage = drainage_systemt[['자치구', '맨홀(개소)', '빗물받이(개소)']]

drainage.set_index('자치구', inplace=True)
drainage.head()
```

## MinMaxScaler()

```python
from sklearn import preprocessing

col = ['맨홀(개소)', '빗물받이(개소)']

x = drainage[col].values
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
drainage = pd.DataFrame(x_scaled, columns = col, index = drainage.index)
drainage.head()
```

```python
import folium
map = folium.Map(location=[37.5502, 126.982], zoom_start=10,
                tiles='openstreetmap')
# tiles='cartodbpositron'
map.choropleth(geo_data=geo_str,
               data = drainage['맨홀(개소)'],
               columns = [drainage.index, drainage['맨홀(개소)']],
               fill_color='PuRd',
               key_on='feature.properties.name',
               highlight=True,
               fill_opacity=0.5, line_opacity=1)
map
```
