```python
# 데이터 정리 
import numpy as np
import pandas as pd
rainfall_amount = pd.read_csv('Seoul_rainfall_amount.csv', encoding='utf-8')

rainfall_amount = rainfall_amount[['관측소명', '평균강수량', '경도', '위도']]

rainfall_amount = rainfall_amount[0:29]
rainfall = rainfall_amount[['관측소명', '평균강수량', '경도', '위도']]
rainfall.head()
```

```python
import folium
import pandas as pd

map = folium.Map(location=[37.5502, 126.982], zoom_start=10,
                tiles='openstreetmap')

for n in rainfall.index:
    folium.CircleMarker([rainfall['위도'][n], rainfall['경도'][n]], 
                        popup = rainfall['관측소명'][n],
                        color='crimson', fill=True, fill_color='crimson', 
                        radius= rainfall['평균강수량'][n]/100,
).add_to(map)
map
```
