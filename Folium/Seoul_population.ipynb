```python
import json
geo_path = 'Seoul2.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
# 데이터 정리 
import numpy as np
import pandas as pd
population = pd.read_csv('Seoul_population.csv', encoding='utf-8', thousands=',')
population.head()
```

```python
population.rename(columns={population.columns[1] : '구별'}, inplace=True)
population = population[1:28]
population
```

```python
population = population[['구별','인구']]
population.set_index('구별', inplace=True)
population.head()
```

```python
from sklearn import preprocessing

col = ['인구']

x = population[col].values
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
population = pd.DataFrame(x_scaled, columns = col, index = population.index)
population.head()
```

```python
import folium
map = folium.Map(location=[37.5502, 126.982], zoom_start=9,
                tiles='cartodbpositron')
map.choropleth(geo_data=geo_str,
               data = population['인구'],
               columns = [population.index, population['인구']],
               fill_color='YlOrRd',
               key_on='feature.properties.name',
               highlight=True,
               fill_opacity=0.5, line_opacity=1)
map.save('Seoul_population_map.html')
```
