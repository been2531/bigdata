## 기후(맑음)에 따른 교통사고 

```python
# 데이터 정리 
import numpy as np
import pandas as pd
weather = pd.read_csv('기상상태별_교통사고_18년.csv', encoding='euc-kr', thousands=',')
weather.head()
```

```python
weather.rename(columns={weather.columns[1] : '구별'}, inplace=True)

occur = weather[3:78:3]
clear_occur = occur[['구별', '구분', '맑음']]
clear_occur = clear_occur.reset_index()
clear_occur = clear_occur[['구별', '구분', '맑음']]

clear_occur.set_index('구별', inplace=True)
clear_occur.head()
```

## MinMaxScaler()

```python
from sklearn import preprocessing

col = ['맑음']

x = clear_occur[col].values
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
clear_occur = pd.DataFrame(x_scaled, columns = col, index = clear_occur.index)
clear_occur.head()
```

## folium

```python
import folium
map = folium.Map(location=[37.5502, 126.982], zoom_start=9,
                tiles='cartodbpositron')
map.choropleth(geo_data=geo_str,
               data = clear_occur['맑음'],
               columns = [clear_occur.index, clear_occur['맑음']],
               fill_color='PuRd',
               key_on='feature.properties.name',
               highlight=True,
               fill_opacity=0.5, line_opacity=1)
map
```
